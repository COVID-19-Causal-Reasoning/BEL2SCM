```python
import pandas as pd
import matplotlib.pyplot as p
```

```python
#reparametrized data
toci_reparam_databel2scm=pd.read_csv("causal_effect_sigmoid_with_estimated_parameters_datapoint1.csv")
toci_reparam_dgscm=pd.read_csv("causal_effect_sigmoid_with_known_parameters_datapoint1.csv")
```

```python
toci_reparam_databel2scm.describe(), toci_reparam_dgscm.describe()
```

```python
p.figure(figsize=(19,19))
p.rcParams["font.weight"] = "bold"
p.rcParams["axes.edgecolor"] = 'black'
p.rcParams["axes.linewidth"] = 5
p.hist(toci_reparam_dgscm['0']*(-1) ,bins=range(-5,10), label='SCM with known weights',color='#3182bd',density=True)
p.hist(toci_reparam_databel2scm['0']*(-1),bins=range(-5,10),
       alpha=0.9,
       label='SCM with estimated weights',color='#dede00',density=True)
p.xlabel("Causal effect of setting Toci to 0 on cytokine storm",fontsize=45,weight='bold')
p.ylabel("Proability density",fontsize=45,weight='bold')
p.legend(loc='upper right',fontsize=50)
p.tick_params(labelsize=35)
p.savefig('toci0_counterfactual.pdf')
```
